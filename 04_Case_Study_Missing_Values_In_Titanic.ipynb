## we have 3 columns which has missing values: Age, Cabin, Embarked
     

## Cabin column

```python
cabin_df = titanic_df['Cabin']
cabin_df.head(10)
```

```python
pcnt = (cabin_df.isnull().sum()/len(titanic_df))*100
pcnt
```

## Over 77% values in this column are missing, its better to drop this column 

```python
titanic_df.drop(columns=['Cabin'], inplace=True)
```

```python
titanic_df
```

## Age 

```python
age_df = titanic_df['Age']
age_df.head(10)
```

```python
age_pct = (age_df.isnull().sum()/len(titanic_df))*100
age_pct
```

```python
titanic_df['Age'].hist(bins=45)
```

```python
age_df = titanic_df[['Age', 'Sex']]
age_df
```

```python
age_df['Age'] = age_df.groupby('Sex').transform(lambda x: x.fillna(x.mean()))
```

```python
age_df.hist(bins=45)
```
