### Pandas -- Data Analysis Tool Kit

Data Structure of Pandas 

1. Series --> one dimensional data structure

   pd.Series(data,index=None)
   
   data can be python data structure or numpy array
   
   
2. DataFrame --> 2 dimensional Data Structure with rows and columns

   pd.DataFrame(data,index=None,columns=None)
   
   data can be list , set, tuple or dictionary or stored array, numpy array etc
   
   index  --> Its a row-label which can be given as per requirement.  (starts with zero)
   
   columns --> its a column-label which can be given as per requirement. (starts with zero)

### Note:
to access elements of dataframe by row-labels as well as column-labelss we use .loc method

```python
df2.loc['20201124':'20201125', ['A', 'B']]
```

```python
df2.loc[::2,::2]
```

### Note

To Access elements of DataFrame by position / index of the Datafrme we used iloc method

```python
df2.iloc[[0,2,4,6],[1,3]]
```

```python
df2.iloc[:3,]
```

```python
df2.iloc[:,:3]
```

###  Differences between iloc and loc
